```python
import pandas as pd

voice_data = pd.read_csv('voice.csv')
```

```python
voice_data.head(10)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
lbe = LabelEncoder()
```

```python
voice_data['label'] = lbe.fit_transform(voice_data['label'])
```

```python
voice_data.head()
```

```python
import lightgbm as lgb
```

```python
param = {'boosting_type':'gbdt',
                         'objective' : 'binary', 
                         'metric' : 'auc',
                         'learning_rate' : 0.01,
                         'max_depth' : 15,
                         'feature_fraction':0.8,
                         'bagging_fraction': 0.9,
                         'bagging_freq': 8,
                         'lambda_l1': 0.6,
                         'lambda_l2': 0,}
```

```python
x_train, x_test, y_train, y_test = train_test_split(voice_data.drop('label', axis=1), voice_data['label']
                                                    , test_size = 0.2)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
x_test.head()
```

```python
y_train.head()
```

```python
train_data = lgb.Dataset(x_train, y_train)
test_data = lgb.Dataset(x_test, y_test)
```

```python
model = lgb.train(param,train_data,valid_sets=[train_data,test_data],num_boost_round = 10000 ,early_stopping_rounds=200,verbose_eval=25)
```
