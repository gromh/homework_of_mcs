```
train = pd.read_csv('Market_Basket_Optimisation.csv', header=None)
```

```
train.head()
```

```
train.shape
```

```
transactions = []
for i in range(0, train.shape[0]):
  temp = []
  for j in range(20):
    if str(train.values[i, j]) != 'nan':
      temp.append(str(train.values[i, j]))
    transactions.append(temp)
```

```
transactions
```

```
itemsets, rules = apriori(transactions, min_support=0.02, min_confidence=0.2)
```
